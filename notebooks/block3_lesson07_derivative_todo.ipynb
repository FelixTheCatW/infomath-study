---

# Что такое производная простыми словами

Производная показывает, **как быстро меняется функция**.

Например:

- если машина едет по дороге, производная пути — это скорость;
- если цена растёт, производная показывает скорость роста цены;
- если ошибка модели уменьшается, производная помогает понять направление улучшения.

---

# Простая идея

Есть функция:

```python
y = x ** 2
```

Если `x` немного изменился, то `y` тоже изменился.

Производная показывает:

```text
насколько сильно меняется y при маленьком изменении x
```

---

# Что такое касательная

Касательная — это прямая, которая касается графика в одной точке и показывает направление движения функции в этой точке.

Производная связана с наклоном касательной.

---

# Что такое дифференцируемость

Функция дифференцируема в точке, если в этой точке у неё есть производная.

Обычно для начинающих можно запомнить так:

если график гладкий и без угла или разрыва, производная обычно существует.

---

# Зачем это нужно для AI

В машинном обучении производная нужна для обучения моделей.

Модель делает ошибку.  
Алгоритм смотрит на производную ошибки.  
Потом двигается в сторону уменьшения ошибки.

Это основа **градиентного спуска**.

---

# Зачем это нужно в дипломе

В дипломном проекте студент может использовать производную для понимания:

- как меняется ошибка модели;
- как работает оптимизация;
- почему модель обучается постепенно;
- почему график ошибки должен снижаться;
- почему важен минимум функции.

````python
x_values = list(np.linspace(2.8, 3.2, 21))
y_values = [x**2 for x in x_values]
table = pd.DataFrame({"x": x_values, "x^2": y_values})
table
````

## Ячейка 4. TODO: численная производная

````python
def f(x):
    return x**2


x0 = 3
h = 0.001
derivative = (f(x0 + h) - f(x0)) / h
print("Приближенная производная в x=3:", derivative)

assert derivative > 5.9 and derivative < 6.1
````

## Ячейка 5. TODO: производная в нескольких точках

````python
def numerical_derivative(func, x, h=0.001):
    return (func(x + h) - func(x)) / h


points = list(np.linspace(-3, 3, 21))

derivative = [numerical_derivative(f, point) for point in points]

table = pd.DataFrame(
    {
        "x": points,
        "Приближенная производая": derivative,
        "Точная производная 2x": [2 * point for point in points],
    }
)

table
````

## Ячейка 6. TODO: график функции и производной

````python
x = np.linspace(-5, 5, 300)

y = x**2
dy = 2 * x

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = x²")
plt.plot(x, dy, label="y' = 2x")

plt.title("Функция и производная")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

## Ячейка 7. TODO: постройте касательную

````python
x = np.linspace(-1, 5, 300)


def f(x):
    return x**2


x0 = 3
y0 = f(x0)
slope = 2 * x0

tangent = slope * (x - x0) + y0

plt.figure(figsize=(8, 5))
plt.plot(x, f(x), label="y = x²")
plt.plot(x, tangent, label="касательная в x=3")
plt.scatter([x0], [y0], label="точка касания")

plt.title("Касательная к графику")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert slope == 6
````

## Ячейка 8. TODO: определите рост и убывание

````python
x_points = [-2, -1, 0, 1, 2]

for point in x_points:
    derivative_value = 2 * point

    if derivative_value > 0:
        message = "функция растёт"
    elif derivative_value < 0:
        message = "функция убывает"
    else:
        message = "точка минимума или горизонтальная касательная"

    print(f"x = {point:2} | производная = {derivative_value:2} | {message}")
````

## Ячейка 9. TODO: производная и ошибка модели

````python
x = np.linspace(-5, 5, 300)

loss = x**2
gradient = 2 * x

plt.figure(figsize=(8, 5))
plt.plot(x, loss, label="loss = x²")
plt.plot(x, gradient, label="gradient = 2x")
plt.scatter([0], [0], label="минимум ошибки")

plt.title("Ошибка модели и производная")
plt.xlabel("параметр модели")
plt.ylabel("значение")
plt.legend()
plt.grid(True)
plt.show()

assert min(loss) >= 0
````
