# Что такое производная простыми словами

Производная показывает, **как быстро меняется функция**.

Например:

- если машина едет по дороге, производная пути — это скорость;
- если цена растёт, производная показывает скорость роста цены;
- если ошибка модели уменьшается, производная помогает понять направление улучшения.

---

# Простая идея

Есть функция:

```python
y = x ** 2
```

Если `x` немного изменился, то `y` тоже изменился.

Производная показывает:

```text
насколько сильно меняется y при маленьком изменении x
```

---

# Что такое касательная

Касательная — это прямая, которая касается графика в одной точке и показывает направление движения функции в этой точке.

Производная связана с наклоном касательной.

---

# Что такое дифференцируемость

Функция дифференцируема в точке, если в этой точке у неё есть производная.

Обычно для начинающих можно запомнить так:

если график гладкий и без угла или разрыва, производная обычно существует.

---

# Зачем это нужно для AI

В машинном обучении производная нужна для обучения моделей.

Модель делает ошибку.  
Алгоритм смотрит на производную ошибки.  
Потом двигается в сторону уменьшения ошибки.

Это основа **градиентного спуска**.

---

# Зачем это нужно в дипломе

В дипломном проекте студент может использовать производную для понимания:

- как меняется ошибка модели;
- как работает оптимизация;
- почему модель обучается постепенно;
- почему график ошибки должен снижаться;
- почему важен минимум функции.
