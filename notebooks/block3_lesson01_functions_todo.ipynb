# Что такое функция

Функция — это правило, которое каждому значению x ставит в соответствие значение y.

Пример:

```python
y = 2 * x + 1
```

Если:

```python
x = 1
```

то:

```python
y = 3
```

## Ячейка 3. TODO: постройте график линейной функции

Постройте график:

```python
y = 2x + 1
```

````python
x = np.linspace(-10, 10, 100)

y = 2 * x + 1

plt.figure(figsize=(8, 5))
plt.plot(x, y)

plt.title("Линейная функция y = 2x + 1")
plt.xlabel("x")
plt.ylabel("x")

plt.grid(True)

plt.show()
````

## Ячейка 4. TODO: сравните несколько линейных функций

Постройте:
- y = x
- y = 2x
- y = -x

````python
x = np.linspace(-10, 10, 100)

y1 = x
y2 = 3 * x
y3 = -x

plt.figure(figsize=(8, 5))

plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = 3x")
plt.plot(x, y3, label="y = -x")

plt.legend()
plt.grid(True)

plt.title("Влияние коэффициента k")

plt.show()
````

## Ячейка 6. TODO: создайте квадратичную функцию

Постройте график:

```python
y = x²
```

````python
x = np.linspace(-10, 10, 100)

y = x ** 2

plt.figure(figsize=(8, 5))

plt.plot(x, y)

plt.title("Квадратичная функция y = x²")
plt.legend()
plt.grid(True)

plt.show()
````

## Ячейка 7. TODO: измените квадратичную функцию

Сравните:
- y = x²
- y = 2x²
- y = x² + 10

````python
x = np.linspace(-10, 10, 100)
y1  = x ** 2
y2 = 2 * x ** 2
y3 = x ** 2 + 10

plt.figure(figsize=(8, 5))

plt.plot(x, y1, label="y = x²")
plt.plot(x, y2, label="y = 2x²")
plt.plot(x, y3, label="y = x² + 10")

plt.legend()
plt.grid(True)

plt.title("Изменение квадратичной функции")

plt.show()
````

## Ячейка 8. TODO: найдите минимум и максимум

Используйте:
- np.min()
- np.max()

````python
x = np.linspace(-5, 5, 100)

y = x ** 2

print("Минимум:", np.min(y))
print("Максимум:", np.max(y))

assert np.min(y) >= 0
````

## Ячейка 9. TODO: постройте график обучения модели

Создайте словарь:
- epoch
- loss

И постройте график.

````python
metrics = {"epoch": [1, 2, 3, 4, 5], "loss": [0.9, 0.7, 0.5, 0.4, 0.3]}

plt.figure(figsize=(8, 5))

plt.plot(metrics["epoch"], metrics["loss"])

plt.title("Пример графика обучения модели")

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

assert len(metrics["epoch"]) == 5
````
