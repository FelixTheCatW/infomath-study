## Ячейка 1. TODO: создайте список чисел

````python
!pip install psycopg2-binary dotenv

import psycopg2
from dotenv import load_dotenv
import os

# conn = psycopg2.connect(
#     host="localhost",
#     database="nitrition_db",
#     user="postgres",
#     password="11",
#     client_encoding="utf8",
# )
load_dotenv()
connection = psycopg2.connect(os.getenv("DATABASE_URI_LOCAL"))

cursor = connection.cursor()

assert connection is not None
assert cursor is not None
print("Соединение установлено, курсор создан")
````

## Ячейка 2. TODO: посчитайте среднее значение

````python
cursor.execute("SELECT calories FROM food")
calories = [c[0] for c in cursor.fetchall()]


average_calories = sum(calories) / len(calories)

print("Средняя калорийность блюда:", round(average_calories, 2))

assert average_calories > 0
````

## Ячейка 3. TODO: найдите максимум и минимум

````python
max_calories = max(calories)
min_calories = min(calories)

print("Максимум:", max_calories)
print("Минимум:", min_calories)

assert max_calories > min_calories > 0
````

## Ячейка 4. TODO: отфильтруйте список

````python
high_calories = []

for c in calories:
    if c > average_calories:
        high_calories.append(c)

print("Калорийность выше средней:", high_calories)
assert all(c > average_calories for c in high_calories)
````

## Ячейка 5. TODO: создайте словарь объекта

````python
import psycopg2.extras

with connection.cursor(cursor_factory=psycopg2.extras.RealDictCursor) as cur:
    cur.execute("SELECT name, calories FROM food LIMIT 1")
    food = cur.fetchall()[0]

print(f"{food["name"]} — {food["calories"]} кКал")
    
assert "Борщ" in food.values()
````

## Ячейка 6. TODO: создайте список словарей

````python
with connection.cursor(cursor_factory=psycopg2.extras.RealDictCursor) as cur:
    cur.execute("SELECT * FROM food")
    foods = cur.fetchall()
    
for food in foods:
    print(f"{food['name']} — {food['calories']} кКал")

assert len(foods) == 163
````

## Ячейка 7. TODO: сделайте поиск по условию

````python
russian_food = [food["name"] for food in foods if food["locale"] == "ru_RU"]

print("Российские блюда:", ", ".join([f.lower() for f in russian_food]))

assert len(russian_food) > 21
````

## Ячейка 8. TODO: отсортируйте список словарей

````python
sorted_food = sorted(foods, key=lambda x: x["protein_g"], reverse=True)

for f in sorted_food[:10]:
    print(f)

assert sorted_food[0]["protein_g"] > sorted_food[-1]["protein_g"]
````

## Ячейка 9. TODO: найдите уникальные значения через set

````python
locales = [f["locale"] for f in foods]

unique_locales = set(locales)

print("Список мест:", unique_locales)

assert len(unique_locales) >= 10
````

## Ячейка 10. TODO: соберите мини-алгоритм для диплома

````python
import random
selected_locale = random.choice(list(unique_locales))

filtered = [f for f in foods if f["locale"] == selected_locale]

sorted_filtered = sorted(filtered, key= lambda x: x["protein_g"], reverse=True)

best_candidate = sorted_filtered[0]

print(f"Самый белковосодержащий продукт в {selected_locale} - {best_candidate["name"]}")

connection.close()
````
