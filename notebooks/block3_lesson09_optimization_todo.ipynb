# Что такое оптимизация

Оптимизация — это поиск лучшего значения.

Обычно мы ищем:

- минимум;
- максимум.

Примеры:

- найти минимальную ошибку модели;
- найти максимальную прибыль;
- найти минимальные затраты;
- найти оптимальное время тренировки;
- найти лучший score кандидата.

---

# Что такое минимум функции

Минимум — это точка, где функция принимает наименьшее значение.

Пример:

```python
f(x) = x²
```

Минимум находится в точке:

```python
x = 0
```

потому что:

```python
f(0) = 0
```

---

# Что такое максимум функции

Максимум — это точка, где функция принимает наибольшее значение.

Например, если есть график прибыли, максимум показывает лучший результат.

---

# Связь с производной

Производная помогает понять, где функция растёт или убывает.

Если производная равна 0, это может быть:

- минимум;
- максимум;
- точка перегиба.

Такие точки называют **критическими точками**.

---

# Зачем это нужно для AI

Машинное обучение — это задача оптимизации.

Модель делает ошибку.  
Нужно найти такие параметры, при которых ошибка минимальна.

Это и есть оптимизация.

---

# Простая идея градиентного спуска

Если мы стоим на графике функции ошибки, производная показывает направление роста.

Чтобы уменьшить ошибку, нужно двигаться в противоположную сторону.

Так работает градиентный спуск.

````python
# TODO: напишите функцию derivative(x) для f(x)=(x-3)**2+2
def derivative(x):
    return 2 * (x - 3)
# Подсказка: производная равна 2 * (x - 3)

# TODO: создайте список points
points = np.linspace(-5, 5, 20)

# TODO: посчитайте derivatives
derivatives = [derivative(x) for x in points]

# TODO: создайте DataFrame
table = pd.DataFrame({
    "x": points,
    "y": derivatives
})
# TODO: покажите таблицу
table
````

````python
# TODO: для каждой точки из points посчитайте производную
points = np.linspace(-10, 10, 20)

for x in points:
    d = derivative(x)
        
# TODO: если производная < 0 — функция убывает

# TODO: если производная > 0 — функция растёт

# TODO: если производная == 0 — критическая точка
    if d > 0:
        message = "↗ функция растёт"
    elif d < 0:
        message = "↘ функция убывает"
    else:
        message = "★ возможная точка экстремума"
# TODO: выведите результат
    print(f"x = {round(x, 2):6}| f'(x) = {round(d, 2):6}|  {message}")
````

## Ячейка 8. TODO: имитация градиентного спуска

Минимизируйте функцию:

```python
f(x) = (x - 3)² + 2
```

Производная:

```python
f'(x) = 2(x - 3)
```

````python
# TODO: напишите функцию f(x)
def f(x):
    return (x-3)**2 + 2
# TODO: напишите функцию df(x)
def df(x):
    return 2 * (x-3)
# TODO: задайте x_current
x_current = -3
# TODO: задайте learning_rate
learning_rate = 0.2
# TODO: создайте history_x и history_y
history_x = []
history_y = []
# TODO: выполните цикл из 15 шагов
for step in range(15):
    history_x.append(x_current)
    history_y.append(f(x_current))

    # TODO: обновляйте x_current по формуле:
    # x_current = x_current - learning_rate * gradient
    
    gradient = df(x_current)
    x_current = x_current - learning_rate * gradient

# TODO: выведите финальное x и f(x)
print("Финальное x:", x_current)
print("Финальное f(x):", f(x_current))
# TODO: assert
assert abs(x_current - 3) < 1
````

## Ячейка 9. TODO: график шагов оптимизации

````python
# TODO: создайте x и y для функции
x = np.linspace(-10, 10, 300)
# TODO: постройте график функции
y = f(x)

# TODO: добавьте точки history_x, history_y
plt.figure(figsize=(8,5))
plt.plot(x, y, label="f(x)=f(x-3)**2 + 2")
plt.scatter(history_x, history_y, label="шаги оптимизации")

plt.title("Градиентный спуск приблежается к минимуму")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# TODO: show()
plt.show()
# TODO: assert history_y[-1] < history_y[0]
assert history_y[-1] < history_y[0]
````
