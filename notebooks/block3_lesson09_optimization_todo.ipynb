# Что такое оптимизация

Оптимизация — это поиск лучшего значения.

Обычно мы ищем:

- минимум;
- максимум.

Примеры:

- найти минимальную ошибку модели;
- найти максимальную прибыль;
- найти минимальные затраты;
- найти оптимальное время тренировки;
- найти лучший score кандидата.

---

# Что такое минимум функции

Минимум — это точка, где функция принимает наименьшее значение.

Пример:

```python
f(x) = x²
```

Минимум находится в точке:

```python
x = 0
```

потому что:

```python
f(0) = 0
```

---

# Что такое максимум функции

Максимум — это точка, где функция принимает наибольшее значение.

Например, если есть график прибыли, максимум показывает лучший результат.

---

# Связь с производной

Производная помогает понять, где функция растёт или убывает.

Если производная равна 0, это может быть:

- минимум;
- максимум;
- точка перегиба.

Такие точки называют **критическими точками**.

---

# Зачем это нужно для AI

Машинное обучение — это задача оптимизации.

Модель делает ошибку.  
Нужно найти такие параметры, при которых ошибка минимальна.

Это и есть оптимизация.

---

# Простая идея градиентного спуска

Если мы стоим на графике функции ошибки, производная показывает направление роста.

Чтобы уменьшить ошибку, нужно двигаться в противоположную сторону.

Так работает градиентный спуск.
