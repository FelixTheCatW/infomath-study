# Главное правило простыми словами

Если функции ведут себя нормально около точки, то предел можно считать по частям.

Например:

```text
lim (f(x) + g(x)) = lim f(x) + lim g(x)
```

Это значит:

если одна часть стремится к 2, а другая к 3, то сумма стремится к 5.

## Ячейка 3. TODO: график суммы функций

````python
x_values = np.linspace(2.9, 3.1, 11)
f_values = list(x_values)
g_values = [3 * x for x in x_values]
h_values = [f + g for f, g in zip(f_values, g_values)]

plt.figure(figsize=(8, 5))
plt.plot(x_values, f_values, label="f(x)=x")
plt.plot(x_values, g_values, label="g(x)=3x")
plt.plot(x_values, h_values, label="h(x)=f(x)+g(x)")
plt.scatter([3], [12], label="x=3, h=6")

plt.title = "Предел суммы функций"
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)


plt.show()

assert h_values[0] == f_values[0] + g_values[0]
````

## Ячейка 6. TODO: неопределённость 0/0

Используйте функцию:

```python
y = (x ** 2 - 9) / (x - 3)
```

При `x → 3` предел должен стремиться к 6.

````python
# TODO: создайте x_values около 3, но без самого 3
x_values = np.linspace(2.1, 3.9, 20)
x_values = x_values[~np.isclose(x_values, 3.0)]

y_values = [(x ** 2 - 9) / (x - 3) for x in x_values]
# TODO: посчитайте y_values
y_values_simple = [x+3 for x in x_values]
# TODO: создайте DataFrame
table = pd.DataFrame({"x": x_values, "y_1": y_values, "y_2": y_values_simple})
# TODO: покажите таблицу

print(table)

assert 3.0 not in x_values
assert np.allclose(y_values, y_values_simple)
````

## Ячейка 8. TODO: сравните исходную и упрощённую функцию

````python
x_values = np.linspace(2.1, 3.9, 20)
x_values = x_values[~np.isclose(x_values, 3.0)]

y_values = [(x**2 - 9) / (x - 3) for x in x_values]

y_values_simple = [x + 3 for x in x_values]

table = pd.DataFrame({"x": x_values, "исходная функция": y_values, "упрошенная функция": y_values_simple})

print(table)

assert 3.0 not in x_values
assert np.allclose(y_values, y_values_simple)
````

## Ячейка 9. TODO: график стабилизации метрики

````python
epochs = [1, 2, 3, 4, 5, 6, 7, 8]
accuracy = np.linspace(0.6, 0.945, 8)

plt.figure(figsize=(8, 5))
plt.plot(epochs, accuracy, marker="o")
plt.axhline(0.95, linestyle="--", label="возможный предел 0.95")

plt.title = "Метрика модели приближается к пределу"
plt.xlabel("Эпоха")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

assert accuracy[-1] > accuracy[0]
````
