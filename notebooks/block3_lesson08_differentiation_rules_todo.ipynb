# Что такое дифференцирование

Дифференцирование — это нахождение производной функции.

Производная показывает, как быстро меняется функция.

---

# Что изучаем сегодня

Сегодня разберём простые правила:

- производная константы;
- производная линейной функции;
- производная степенной функции;
- производная суммы;
- производная функции с коэффициентом;
- сравнение функции и производной на графике.

---

# Главное правило простыми словами

Если функция простая, её производную можно найти по правилам.

Примеры:

```text
f(x) = 5        → f'(x) = 0
f(x) = x        → f'(x) = 1
f(x) = x²       → f'(x) = 2x
f(x) = x³       → f'(x) = 3x²
f(x) = 3x²      → f'(x) = 6x
```

---

## Ячейка 1. TODO: импортируйте библиотеки

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Библиотеки подключены")
````

## Ячейка 2. TODO: производная константы

````python
x = np.linspace(-8, 8, 200)

f = np.ones_like(x) * 8
df = np.zeros_like(x)

plt.figure(figsize=(8, 5))
plt.plot(x, f, label="f(x)=8")
plt.plot(x, df, label="f'(x)=0")

plt.title("Производная константы")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert np.all(df == 0)
````

## Ячейка 3. TODO: производная линейной функции

````python
x = np.linspace(-5, 5, 200)

f = 3 * x + 1
df = np.ones_like(x) * 3

plt.figure(figsize=(8, 5))
plt.plot(x, f, label="f(x)=3x+1")
plt.plot(x, df, label="f'(x)=3")

plt.title("Производная линейной функции")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert df[0] == 3
````

## Ячейка 4. TODO: производная x²

````python
x = np.linspace(-5, 5, 200)

f = x**2
df = 2 * x

plt.figure(figsize=(8, 5))
plt.plot(x, f, label="f(x)=x^2")
plt.plot(x, df, label="f'(x)=2x")

plt.title("Производная функции x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert len(f) == len(df)
````

## Ячейка 5. TODO: производная x³

````python
x = np.linspace(-5, 5, 200)

f = x**3
df = 3 * x**2

plt.figure(figsize=(8, 5))
plt.plot(x, f, label="f(x)=x^4")
plt.plot(x, df, label="f'(x)=3x^3")

plt.title("Производная функции x^3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert np.min(df) >= 0
````

## Ячейка 6. TODO: производная функции с коэффициентом

````python
x = np.linspace(-5, 5, 200)

f = 4 * x**2
df = 8 * x

plt.figure(figsize=(8, 5))
plt.plot(x, f, label="f(x)=4x^2")
plt.plot(x, df, label="f'(x)=8x")

plt.title("Производная функции с коэффициентом")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert len(f) == len(df)
````

## Ячейка 7. TODO: производная суммы

````python
x = np.linspace(-5, 5, 200)

f = x**2 + 3 * x
df = 2 * x + 3

plt.figure(figsize=(8, 5))
plt.plot(x, f, label="f(x)=x^2+3x")
plt.plot(x, df, label="f'(x)=2x+2")

plt.title("Производная суммы")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert len(f) == len(df)
````

## Ячейка 9. TODO: рост и убывание функции

````python
# TODO: для каждой точки из points посчитайте derivative = 2*x + 3

points = np.linspace(-5, 5, 21)

for x in points:
    derivative = 2 * x + 3

    if derivative > 0:
        message = "↗ функция растёт"
    elif derivative < 0:
        message = "↘ функция убывает"
    else:
        message = "★ возможная точка экстремума"

    print(f"x = {round(x,2):5}| f'(x) = {round(derivative, 2):5}|  {message}")
````
