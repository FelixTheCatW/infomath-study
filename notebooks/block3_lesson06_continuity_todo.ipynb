# Что такое непрерывность простыми словами

Функция непрерывна, если её график можно нарисовать “не отрывая карандаш от бумаги”.

Простой пример непрерывной функции:

```python
y = x ** 2
```

Её график плавный, без разрывов и скачков.

---

# Что такое разрыв

Разрыв — это место, где график:

- обрывается;
- имеет дырку;
- резко перескакивает;
- не определён в точке.

Примеры разрывов:

- деление на ноль;
- функция с условием;
- резкий скачок значения.

---

# Почему непрерывность важна

Если функция непрерывна, с ней проще работать:

- строить графики;
- анализировать поведение;
- искать минимумы и максимумы;
- использовать методы оптимизации.

---

# Зачем это нужно для AI

В машинном обучении часто используются функции ошибки.

Модель обучается, когда постепенно уменьшает ошибку.

Если функция ошибки ведёт себя плавно, её проще оптимизировать.

Поэтому непрерывность связана с:
- производной;
- градиентным спуском;
- оптимизацией;
- обучением моделей.

---

# Зачем это нужно в дипломе

- почему гладкие графики проще анализировать;
- почему резкие скачки могут быть проблемой;
- почему данные и функции нужно визуализировать;
- как объяснить поведение модели или метрики.

## Ячейка 2. TODO: постройте непрерывную функцию y = x²

````python
x = np.linspace(-5, 5, 300)
y = x**2 + 4 * x + 2

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title = "Непрерывная функция y = x² + 4x + 2"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

assert len(x) == len(y)
````

## Ячейка 3. TODO: постройте непрерывную функцию y = cos(x)

````python
x = np.linspace(-2 * np.pi, 2 * np.pi, 300)
y = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title ="Непрерывная функция y = cos(x)"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)


assert np.max(y) <= 1.01
assert np.min(y) >= -1.01
````

## Ячейка 4. TODO: постройте функцию с разрывом y = 1/x

````python
x_left = np.linspace(-5, -0.1, 300)
x_right = np.linspace(0.1, 5, 300)

y_left = 1 / x_left
y_right = 1 / x_right

plt.figure(figsize=(8, 5))
plt.plot(x_left, y_left)
plt.plot(x_right, y_right)

plt.title = "Функция с разрывом y = 1/x"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

## Ячейка 6. TODO: функция с дыркой

Используйте функцию:

```python
y = (x ** 2 - 4) / (x - 2)
```

Она не определена при `x = 2`, но около этой точки ведёт себя как `x + 2`.

````python
# TODO: создайте x_left до 2

# TODO: создайте x_right после 2
x_left = np.linspace(-2, 1.99, 200)
x_right = np.linspace(2.01, 6, 200)

y_left = (x_left**2 - 4) / (x_left - 2)
y_right = (x_right**2 - 4) / (x_right - 2)

plt.figure(figsize=(8, 5))
plt.plot(x_left, y_left)
plt.plot(x_right, y_right)
plt.scatter([2], [4], facecolors="none", edgecolors="black", label="дырка")

plt.title = "Функция с дыркой"
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

assert 4 not in y_right or 4 not in y_left
````

## Ячейка 7. TODO: создайте функцию со скачком

````python
x = np.linspace(-5, 5, 300)

y = []

for value in x:
    if value < 0:
        y.append(-2)
    else:
        y.append(2)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title = "Функция со скачком"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

assert min(y) == -2
assert max(y) == 2
````

## Ячейка 8. TODO: сравните непрерывную и разрывную функцию

````python
x = np.linspace(-5, 5, 300)

y_smooth = x**2

y_jump = []
for value in x:
    if value < 0:
        y_jump.append(-1)
    else:
        y_jump.append(1)

plt.figure(figsize=(8, 5))
plt.plot(x, y_smooth, label="непрерывная: x^2")
plt.plot(x, y_jump, label="разрывная: скачок")

plt.title = "Сравнение непрерывной и разрывной функции"
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

## Ячейка 9. TODO: графики плавной и скачущей ошибки модели

````python
epochs = [1, 2, 3, 4, 5, 6, 7]

smooth_loss = [0.9, 0.75, 0.6, 0.48, 0.38, 0.31, 0.27]
jump_loss = [0.9, 0.5, 0.8, 0.4, 0.7, 0.3, 0.6]

plt.figure(figsize=(8, 5))
plt.plot(epochs, smooth_loss, marker="o", label="плавное обучение")
plt.plot(epochs, jump_loss, marker="o", label="скачки ошибки")

plt.title = "Плавное и нестабильное изменение ошибки"
plt.xlabel("Эпоха")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

assert smooth_loss[-1] < smooth_loss[0]
````
