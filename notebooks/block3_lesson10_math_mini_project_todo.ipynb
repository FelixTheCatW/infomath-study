## Ячейка 1. TODO: импортируйте библиотеки

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Библиотеки подключены")
````

## Ячейка 2. TODO: создайте функцию ошибки

Используйте функцию:

```python
loss(x) = (x - 4)² + 1
```

Минимум находится при `x = 4`, значение минимума равно `1`.

````python
def loss_function(x):
    return (x - 4) ** 2 + 1


x0 = 4
minimum_value = loss_function(x0)

print("Минимум находится при x =", x0)
print("Минимальное значение loss =", minimum_value)

assert minimum_value == 1
````

## Ячейка 3. TODO: постройте график функции ошибки

````python
# TODO: создайте x через np.linspace()

# TODO: создайте loss = loss_function(x)

# TODO: постройте график

# TODO: отметьте минимум x=4, y=1

# TODO: добавьте title, xlabel, ylabel, legend, grid

# TODO: show()

# TODO: assert

x = np.linspace(-5, 5, 300)
loss = loss_function(x)

idx_min = np.argmin(loss)
x_min = x[idx_min]
loss_min = loss[idx_min]

plt.figure(figsize=(8, 5))
plt.plot(x, loss, label="loss(x) = (x - 4)² + 1")
plt.scatter([x_min], [loss_min], label="минимум")

plt.title("Функция ошибки модели")
plt.xlabel("x")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()

assert min(loss) >= 1
````

## Ячейка 4. TODO: создайте производную функции ошибки

Для функции:

```python
loss(x) = (x - 4)² + 1
```

производная:

```python
loss'(x) = 2(x - 4)
```

````python
# TODO: напишите функцию loss_derivative(x)
def loss_derivative(x):
    return 2 * (x - 4)


# TODO: создайте список points
points = np.linspace(-10, 10, 20)
# TODO: посчитайте derivative_values
derivative_values = loss_derivative(points)
# TODO: создайте DataFrame со столбцами x, loss(x), loss'(x)
df = pd.DataFrame({"points": points, "loss'(x)": derivative_values})
# TODO: покажите таблицу
df
````

## Ячейка 5. TODO: проанализируйте рост и убывание

````python
# TODO: создайте пустой список analysis
def direction(v):
    if v > 0:
        return "↗ функция растёт"
    elif v < 0:
        return "↘ функция убывает"
    else:
        return "★ возможная точка экстремума"

analysis = [
    {
        "x": x,
        "derivative": d,
        "status": direction(d),
    }
    for x, d in zip(points, derivative_values)
]
# TODO: выведите analysis
for i, item in enumerate(analysis, 1):
    print(f"{i:2}  {item['x']:8.4f}  {item['derivative']:12.4f}  {item['status']}")
# TODO: assert
assert len(analysis) == len(points)
````

## Ячейка 6. TODO: сделайте простую оптимизацию

````python
# TODO: задайте x_current далеко от минимума, например -2
x_current = -2
# TODO: задайте learning_rate = 0.2
learning_rate = 0.2
# TODO: задайте steps = 20
steps = 20
# TODO: создайте пустой список history
history = []
# TODO: в цикле сохраните step, x, loss, derivative
for step in range(steps):
    current_loss = loss_function(x_current)
    current_derivative = loss_derivative(x_current)
    
    history.append(
        {
            "step": step,
            "x": x_current,
            "loss": current_loss,
            "derivative": current_derivative,
        }
    )
    

# TODO: обновите x_current по формуле:
# x_current = x_current - learning_rate * derivative
    x_current = x_current - learning_rate * current_derivative
# TODO: выведите финальное x и финальный loss
print("Финальное x:", x_current)
print("Финальный loss:", loss_function(x_current))

assert abs(x_current - 4) < 0.3
````

## Ячейка 7. TODO: создайте таблицу шагов оптимизации

````python
history_df = pd.DataFrame(history)

history_df.head(10)

assert len(history_df) == steps
````

## Ячейка 8. TODO: постройте график движения к минимуму

````python
x = np.linspace(-2, 8, 300)
loss = loss_function(x)

plt.figure(figsize=(8, 5))
plt.plot(x, loss, label="loss(x)")
plt.scatter(history_df["x"], history_df["loss"], label="шаги оптимизации")
plt.scatter([3], [2], label="истинный минимум")

plt.title("Оптимизация функции ошибки")
plt.xlabel("x")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()

assert history_df["loss"].iloc[-1] < history_df["loss"].iloc[0]
````

## Ячейка 9. TODO: создайте итоговый отчёт проекта

````python
project_report = {
    "project_name": "Анализ функции ошибки модели",
    "start_x": history_df["x"].iloc[0],
    "final_x": history_df["x"].iloc[-1],
    "start_loss": history_df["loss"].iloc[0],
    "final_loss": history_df["loss"].iloc[-1],
    "steps": steps,
    "minimum_x": 3,
    "minimum_loss": 2,
}

for key, value in project_report.items():
    print(key, ":", value)

assert project_report["final_loss"] < project_report["start_loss"]
assert history_df["loss"].iloc[-1] < history_df["loss"].iloc[0]
````
