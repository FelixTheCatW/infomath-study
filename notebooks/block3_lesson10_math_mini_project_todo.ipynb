# Зачем это нужно для AI

Машинное обучение — это поиск минимальной ошибки.

Модель обучается так:

1. считает ошибку;
2. смотрит, куда ошибка растёт;
3. двигается в обратную сторону;
4. постепенно приближается к минимуму.

Это основа градиентного спуска.
