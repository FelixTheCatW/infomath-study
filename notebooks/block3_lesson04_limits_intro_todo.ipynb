# Что такое предел простыми словами

Предел показывает, к какому значению стремится функция, когда `x` приближается к некоторому числу.

Пример:

```text
x приближается к 2
```

Мы смотрим:

```text
к чему приближается y
```

---

# Жизненная аналогия

Представьте, что человек идёт к двери.

Он может быть:
- в 10 метрах;
- в 5 метрах;
- в 1 метре;
- в 0.1 метра.

Он всё ближе и ближе к двери.

Предел — это идея “к чему мы приближаемся”.

---

# Пример функции

```python
y = 2 * x
```

Если `x` приближается к 3, то `y` приближается к 6.

То есть:

```text
при x → 3, y → 6
```

---

# Зачем пределы нужны в математике

Пределы нужны, чтобы понять:

- непрерывность;
- производную;
- скорость изменения функции;
- оптимизацию;
- обучение моделей.

---

# Зачем это нужно для AI

Машинное обучение использует оптимизацию.

А оптимизация связана с производной.

А производная строится на идее предела.

Поэтому пределы — это первый шаг к пониманию того, как обучается модель.

---

# Зачем это нужно в дипломе

В дипломе студент может не считать сложные пределы вручную, но должен понимать:

- почему график ведёт себя так или иначе;
- что значит “функция стремится”;
- как модель уменьшает ошибку;
- почему производная важна для обучения AI.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Библиотеки подключены")
````

## Ячейка 2. TODO: создайте таблицу значений для y = 3x

````python
x_values = [2.5, 2.9, 2.99, 3.0, 3.01, 3.1, 3.5]

y_values = []

for x in x_values:
    y_values.append(2 * x)

table = pd.DataFrame({"x": x_values, "y = 2x": y_values})

table
````

## Ячейка 3. TODO: постройте график y = 3x

````python
x = np.linspace(0, 6, 100)
y = 2 * x

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = 2x")
plt.scatter([3], [6], label="точка x=3, y=6")

plt.title = "Предел на примере y = 2x"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

assert 2 * 3 == 6
````

## Ячейка 4. TODO: сделайте приближение слева и справа

````python
left_x = [2.9, 2.99, 2.999]
right_x = [3.1, 3.01, 3.001]

left_y = [2 * x for x in left_x]
right_y = [2 * x for x in right_x]

table = pd.DataFrame(
    {"x слева": left_x, "y слева": left_y, "x справа": right_x, "y справа": right_y}
)

table
````

## Ячейка 6. TODO: постройте график y = x²

````python
x = np.linspace(-1, 5, 200)
y = x**2

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = x²")
plt.scatter([2], [4], label="x=2, y=4")

plt.title = "Предел на примере y = x²"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

assert 2**2 == 4
````

## Ячейка 8. TODO: постройте график функции с “дыркой”

````python
# TODO: создайте x_left до 2
x_left = np.linspace(1, 1.99, 100)

# TODO: создайте x_right после 2
x_right = np.linspace(2.01, 3, 100)

# TODO: посчитайте y_left и y_right
y_left = (x_left**2 - 4) / (x_left - 2)
y_right = (x_right**2 - 4) / (x_right - 2)
# TODO: постройте график
plt.figure(figsize=(8, 5))
plt.plot(x_left, y_left)
plt.plot(x_right, y_right)
plt.scatter([2], [4], facecolors='none', edgecolors='red', label='предел 4')

plt.title = "Функция не проеделна в x=2, но предел есть"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
````

## Ячейка 9. TODO: постройте график уменьшения ошибки модели

````python
epochs = [1, 2, 3, 4, 5, 6, 7]
loss = [0.9, 0.65, 0.45, 0.30, 0.20, 0.14, 0.10]

plt.figure(figsize=(8, 5))
plt.plot(epochs, loss, marker="o")

plt.title = "Ошибка модели стремится к 0"
plt.xlabel("Эпоха")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

assert loss[-1] < loss[0]
````
