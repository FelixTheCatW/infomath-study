## Ячейка 1. TODO: импортируйте библиотеки

````python
import numpy as np

import matplotlib.pyplot as plt

print("Библиотеки импортированы")
````

## Ячейка 2. TODO: постройте график y = x²

````python
x = np.linspace(-10, 10, 100)

y = x ** 2

plt.figure(figsize=(8, 6))

plt.plot(x, y, label="y = x")

plt.legend()
plt.grid(True)

plt.title = "Квалратичная функция"
plt.xlabel("x")
plt.ylabel("y = x ** 2")

plt.show()
print(len(x))
assert len(x) == 100
````

## Ячейка 3. TODO: постройте график y = x³

````python
x = np.linspace(-10, 10, 100)

y = pow(x, 3)

plt.figure(figsize=(8, 3))

plt.plot(x, y)

plt.title = "Кубическая функция y = x³"
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)

plt.show()


assert y[0] < 0
assert y[-1] > 0
````

## Ячейка 4. TODO: постройте график корня

````python
x = np.linspace(0, 25, 200)

y = np.sqrt(x)

plt.figure(figsize=(8, 5))

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.title = "Показательная функция y = √x"
plt.show()

assert np.all(y >= 0)
````

## Ячейка 5. TODO: постройте показательную функцию

````python
x = np.linspace(-5, 5, 200)

y = 2 ** x

plt.figure(figsize=(8, 5))

plt.plot(x, y, label="y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.title = "Кубическая функция y = x³"
plt.show()


assert np.all(y > 0)
````

## Ячейка 6. TODO: постройте логарифмическую функцию

````python
x = np.linspace(0.1, 20, 200)
y = np.log(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title = "Логарифмическая функция y =  log(x)"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

assert len(x) == len(y)
````

## Ячейка 7. TODO: постройте sin(x)

````python
x = np.linspace(-2 * np.pi, 2 * np.pi, 300)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title = "Функция y = sin(x)"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

assert np.max(y) <= 1.01
assert np.min(y) >= -1.01
````

## Ячейка 8. TODO: постройте cos(x)

````python
x = np.linspace(-2 * np.pi, 2 * np.pi, 300)
y = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title = "Функция y = cos(x)"
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

assert np.max(y) <= 1.01
assert np.min(y) >= -1.01
````

## Ячейка 9. TODO: сравните sin(x) и cos(x)

````python
x = np.linspace(-2 * np.pi, 2 * np.pi, 300)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.title = "Сравнение sin(x) и cos(x)"
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

assert len(y_sin) == len(y_cos)
````

## Ячейка 10. TODO: сравните несколько функций

````python
x = np.linspace(0.1, 5, 200)

y_linear = x
y_square = x**2
y_log = np.log(x)
y_exp = 2**x

plt.figure(figsize=(9, 6))
plt.plot(x, y_linear, label="y = x")
plt.plot(x, y_square, label="y = x²")
plt.plot(x, y_log, label="y = log(x)")
plt.plot(x, y_exp, label="y = 2ˣ")

plt.title = "Сравнение элементарных функций"
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

summary = [
    "Степенные функции могут быстро расти",
    "Показательная функция растёт очень быстро",
    "Логарифм растёт медленно",
    "Синус и косинус описывают колебания",
]

for item in summary:
    print("-", item)

assert len(summary) == 4
````
