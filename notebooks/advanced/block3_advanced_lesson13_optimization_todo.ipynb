## Теория

Оптимизация — поиск лучшего решения.

В анализе данных это означает:
- найти максимум;
- найти минимум;
- найти лучший объект по показателю;
- найти худший объект по показателю.

## Практическое применение

- самый прибыльный товар;
- самая эффективная тренировка;
- самый калорийный продукт;
- самый дорогой объект недвижимости.

## Ячейка 1. TODO: импорт библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Ячейка 2. TODO: загрузите датасет

```python
url = "../data/Food_Nutrition_Dataset.csv"

df = pd.read_csv(url)

print("Первые строки датасета:")
display(df.head())

assert len(df) > 0
```

## Ячейка 3. TODO: выделите числовые признаки

```python
numeric_df = df.select_dtypes(include=["number"])
print(numeric_df.columns.tolist())
```

## Ячейка 4. TODO: найдите максимум

```python
max_row = df.loc[df["calories"].idxmax()]
print(max_row[['food_name', 'calories']])
```

## Ячейка 5. TODO: найдите минимум

```python
min_row = df.loc[df['calories'].idxmax()]
print(min_row[['food_name', 'calories']])
```

## Ячейка 6. TODO: постройте гистограмму

```python
plt.hist(df['calories'], bins=20)
plt.title('Распределение калорийности')
```

## Ячейка 7. TODO: получите TOP-10

```python
top10 = df.sort_values('calories', ascending=False).head(10)
display(top10[['food_name', 'calories']])
```

## Ячейка 8. TODO: постройте bar chart

```python
plt.figure(figsize=(8, 5))
plt.bar(top10['food_name'], top10['calories'])
plt.xticks(rotation=90)
```

## Ячейка 9. TODO: найдите среднее

```python
mean_value = df["calories"].mean()
print(mean_value)
assert mean_value > 0
```

## Ячейка 10. TODO: сформулируйте выводы и добавьте assert

```python
print("Лучший штат:", max_row["food_name"])
print("Худший штат:", min_row["food_name"])
print("Среднее:", mean_value)
assert max_row["calories"] >= mean_value
```
